```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Set Values for each Column
total_schools = school_data_complete["school_name"].nunique()
total_students = school_data_complete["student_name"].count()
total_budget = school_data["budget"].sum()
avg_reading = school_data_complete["reading_score"].mean()
avg_math = school_data_complete["math_score"].mean()                                                
passing_math = ((student_data["math_score"]>70).sum() / student_data["math_score"].count())* 100    
passing_reading = ((student_data["reading_score"]>70).sum() / student_data["reading_score"].count())* 100
overall_pass_rate = (school_data_complete["math_score"].mean() + school_data_complete["reading_score"].mean())/2

# # Plug in each column to Data Frame
district_summary = pd.DataFrame({"Total Schools": [total_schools],
          "Total Students": [total_students],
          "Total Budget": [total_budget],
          "Average Math Score": [avg_math],
          "Average Reading Score": [avg_reading],
          "% of Passing Math Scores": [passing_math],
         "% of Passing Reading Scores": [passing_reading],
         "% Overall Passing Rate": [overall_pass_rate]})
district_summary


```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Create Passing Math Group
passing_math_df = school_data_complete.loc[(school_data_complete
                                             ["math_score"] > 70)]
school_math_index = passing_math_df.set_index(["school_name"])
passing_math_group = school_math_index.groupby(["school_name"])

# Create Passing Reading Group
passing_reading_df = school_data_complete.loc[(school_data_complete
                                             ["reading_score"] > 70)]
school_reading_index = passing_reading_df.set_index(["school_name"])
passing_reading_group = school_reading_index.groupby(["school_name"])

# Budget
school_budget = school_data_complete.groupby(["school_name"])["budget"].mean()
```

```python
# Group by School name
school_summary_index_df = school_data_complete.set_index(["school_name"])
school_group = school_summary_index_df.groupby(["school_name"])

# Set Values for each Column
school_types = school_data.set_index(["school_name"])["type"]
total_students = school_group["student_name"].count()
school_budget = school_data_complete.groupby(["school_name"])["budget"].mean()
per_student_budget = school_budget/ school_group["student_name"].count()
avereage_math_score = school_group["math_score"].sum() / school_group["student_name"].count()
avereage_reading_score = school_group["reading_score"].sum() / school_group["student_name"].count() 
percent_passing_math = passing_math_group["math_score"].count() / school_group["math_score"].count()* 100 
percent_passing_reading = passing_reading_group["reading_score"].count() / school_group["reading_score"].count()* 100 
overall_passing_rate = (passing_math_group["math_score"].mean() + passing_reading_group["reading_score"].mean())/2

# Plug in each column to Data Frame
school_summary = pd.DataFrame({"School Type": school_types,
    "Total Students": total_students,
    "Total Budget": school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": avereage_math_score,
    "Average Reading Score": avereage_reading_score,
    "% of Passing Math Scores": percent_passing_math,
    "% of Passing Reading Scores": percent_passing_reading,
    "% Overall Passing Rate": overall_passing_rate})
school_summary
```

## Math Scores by Grade

  * Create a pandas series for each grade. Hint: use a conditional statement.

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
pd.cut(school_summary["Per Student Budget"], spending_bins, labels=group_names).head()
```

```python
#  Add column for Student spending
school_summary["Per Student Spending"] = pd.cut(school_summary["Per Student Budget"], spending_bins, labels=group_names)
# Remove unwanted columns
school_summary_group = school_summary.drop(columns =["Total Students", "Total Budget", "Per Student Budget"])
# Group by Per student spending
sort_by_spending_per_student = school_summary_group.groupby(["Per Student Spending"])
sort_by_spending_per_student.mean()
```

```python
# Observably, as schools spent more money, the test scores seem to decline.
```

## Scores by School Size

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
pd.cut(school_summary["Total Students"], size_bins, labels=group_names).head()
```

```python
#  Add column fo Student spending
school_summary["Total Students in School"] = pd.cut(school_summary["Total Students"], size_bins, labels=group_names)
# Remove unwanted columns
school_summary_size = school_summary.drop(columns =["Per Student Spending","Total Budget", "Per Student Budget"])
# Group by Student size
sort_by_school_size = school_summary_size.groupby(["Total Students in School"])
sort_by_school_size.mean()
```

```python
# The larger the school the more we see lower scores. This may have something to do with the amount of attention a student might get. 
```
